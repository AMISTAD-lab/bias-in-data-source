```python
# Data preprocessing

import csv
from hypothesis_test import *
from kardis_test import *
from mg_calculator import *
from q_finder import *
from graph_distributions import *

file = open(r"Datasets\compas.csv", mode='r', encoding='utf-8-sig')
csvreader = csv.reader(file)
rowlist = [row for row in csvreader]
for category in rowlist[0]:
    index = rowlist[0].index(category)
    exec(category + " = [row[index] for row in rowlist[1:]]")
file.close()

redec = [score_text[x] for x in range(1,len(score_text), 3)]
compnames = [first[x] + " " + last[x] for x in range(1, len(first), 3)]

file = open(r"Datasets\people.csv", mode='r', encoding='utf-8-sig')
csvreader = csv.reader(file)
rowlist = [row for row in csvreader]
for category in rowlist[0]:
    index = rowlist[0].index(category)
    exec(category + " = [row[index] for row in rowlist[1:]]")
file.close()

races = [race[name.index(x)] for x in compnames]
```

```python
# Setup and Experiments
racescore = list(filter(lambda x: not x[1] == 'N/A', list(zip(races,redec))))
whitecount = len(list(filter(lambda x: x[0] == 'Caucasian', racescore)))
whiteprops = [racescore.count(('Caucasian','High'))/whitecount, \
                racescore.count(('Caucasian', 'Medium'))/whitecount, \
                racescore.count(('Caucasian', 'Low'))/whitecount]
AAscore = list(filter(lambda x: x[0] == 'African-American', racescore))

print("Null hypothesis (proposed distribution): " + str(whiteprops) + "\n")
q_whiteprops = hypothesis_test(AAscore,\
                [('African-American','High'),('African-American','Medium'),('African-American','Low')],\
                alpha = 0.05, hypothesis = whiteprops)[4]

graph_distributions(q_whiteprops,\
                    [('African-American','High'),('African-American','Medium'),('African-American','Low')],\
                    hypothesis=whiteprops, filename=r"Results\test1.pdf")
```

```python
print("Null hypothesis (proposed distribution): " + str(3*[1/3]) + "\n")

q_uni = hypothesis_test(AAscore,\
                [('African-American','High'),('African-American','Medium'),('African-American','Low')])[4]

graph_distributions(q_uni,\
                    [('African-American','High'),('African-American','Medium'),('African-American','Low')],\
                    hypothesis=3*[1/3], filename=r"Results\test2.pdf")
```
